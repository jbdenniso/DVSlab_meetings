```python
# All the stuff to import and make some nicknames
import seaborn as sns
# Have any of you seen The West Wing --. Samuel Norman Seaborn
# I have a bad habit of making this sb. This makes your code tough to read for others
import pandas as pd
# this handles our data in dataframes
import matplotlib.pyplot as plt
# not 100% necessary but what I do
plt.rcParams['figure.dpi'] = 150
%matplotlib inline


```

```python
mri=sns.load_dataset('fmri')
# "named" arguments are typically strings
print(type(mri))
#Hey this is a pandas dataframe --> seaborn likes pandas dataframes
#Let's get a feel for this data(
display(mri.head())
mri.groupby('subject').mean()
print(mri.subject.unique())
```

```python
my_IV='subject'
#we have a couple of IV's to look at now
boxes=sns.boxplot(x=my_IV,y='signal',data=mri)
swarms=sns.catplot(x=my_IV,y='signal',data=mri)
swarms=sns.catplot(x=my_IV,y='signal',data=mri,kind='violin')
```

```python
one=sns.relplot(x="timepoint", y="signal", kind="line", data=mri)
two=sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=mri);
three=sns.relplot(x="timepoint", y="signal", hue="region", style="event", kind="line", data=mri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=mri.query("event == 'stim'"));
```
